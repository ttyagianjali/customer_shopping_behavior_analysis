```python
df["Review Rating"] = (df
                            .groupby("Category")["Review Rating"]
                            .transform(lambda x: x.fillna(x.median()))
)
```

```python
df.isnull().sum()
```

```python
df.columns = (
    df.columns
    .str.replace(" ", "_")
    .str.lower()
    .str.strip()
    .str.replace("(", "")
    .str.replace(")", "")
)

df = df.rename(
    columns={
       "purchase_amount_usd" : "purchase_amount"
    }
)
```

```python
#create a new column age_group

df["age_group"] = pd.qcut(df["age"], q = 4, labels = ["Young Adults", "Adults", "Middle Aged", "Old"])

```

```python
frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Annually": 365,
    "Quarterly": 90,
    "Monthly": 30,
    "Bi-Weekly": 14,
    "Every 3 Months": 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df["frequency_of_purchases"].unique()
```

```python
df[["subscription_status", "discount_applied", "promo_code_used"]] = (
    df[["subscription_status", "discount_applied", "promo_code_used"]].replace({"Yes" : 1, "No": 0})
)
```
